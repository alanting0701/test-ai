```python
import pandas as pd

dataFrame = pd.read_csv('world.csv')
dataFrame.info()
```

```python
allDataFrame = dataFrame.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
mask = allDataFrame["洲名"] == "亞洲"
asiaDataFrame = allDataFrame[mask]
asiaDataFrame
```

```python
# 選取台灣資料
mask_taiwan = allDataFrame["國家"] == "台灣"
taiwanDataFrame = allDataFrame[mask_taiwan]
taiwanDataFrame
```

```python
# 選取台灣資料同時選取日本資料
mask_taiwan_japan = (allDataFrame["國家"] == "台灣") | (allDataFrame["國家"] == "日本")
taiwan_japan_DataFrame = allDataFrame[mask_taiwan_japan]
taiwan_japan_DataFrame
```

```python
# 台灣與日本的新增確診數大於20000
mask_cases = (taiwan_japan_DataFrame["新增確診數"] > 20000)
high_cases_taiwan_japan = taiwan_japan_DataFrame[mask_cases]
high_cases_taiwan_japan
```
