```python
# 使用list的comprehension建立

print([i for i in range(1,10)])
print([i for i in range(1,10) if i % 2 == 1])
```

```python
# 使用dict的comprehension建立

word = 'letters'
print({letter: word.count(letter) for letter in word})
print({letter: word.count(letter) for letter in word if letter == 'e'})
```
