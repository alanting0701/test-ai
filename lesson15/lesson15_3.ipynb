```python
import numpy as np

students = np.random.randint(50, high=101, size=(50,5))
students.shape
```

```python
#ndarray的實體方法
students.sum(axis=1)  #每個學生的總分
students.mean(axis=1) #每個學生的平均分
```

```python
#numpy內建的函式

np.sum(students, axis=1)  #每個科目的總分
np.mean(students, axis=1) #每個科目的平均分
```

```python
#1家分店30天的營收

oneStore = np.random.randint(100, 10000, size=(30,))
oneStore
```

```python
#累計每天這家店的營收
import matplotlib.pyplot as plt

oneStoreCum = oneStore.cumsum()
oneStoreCum

figure = plt.figure(figsize=(10,5))
axes = figure.add_subplot(1,1,1)
axes.plot(range(1,31), oneStoreCum,'r-')
axes.set_title("one Store shop sumsum")
axes.set_xlabel("day")
axes.set_ylabel("money")
plt.show()
```

```python
#5家店30天的營收

fiveStore = np.random.randint(100, 10000, size=(5,30))
fiveStore
```

```python
#累計每天5家店的營收

fiveStoreCum = fiveStore.cumsum(axis=1)
fiveStoreCum
```

```python
#累計每天5家店的營收
import matplotlib.pyplot as plt


figure = plt.figure(figsize=(10,5))
axes = figure.add_subplot(1,1,1)
axes.plot(range(1,31), fiveStoreCum[0,:],'r-')
axes.plot(range(1,31), fiveStoreCum[1,:],'y-')
axes.plot(range(1,31), fiveStoreCum[2,:],'g-')
axes.plot(range(1,31), fiveStoreCum[3,:],'m-')
axes.plot(range(1,31), fiveStoreCum[4,:],'c-')
axes.set_title("five Store shop sumsum")
axes.set_xlabel("day")
axes.set_ylabel("money")
plt.show()
```
