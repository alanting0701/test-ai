```python
import pandas as pd
world_df = pd.read_csv("world.csv")
print(world_df.columns)
print(world_df.info())
```

```python
world_df1 = world_df.reindex(columns=['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數'])
world_df1
```

```python
# 使用布林遮罩篩選出「國家」欄位為「台灣」的資料

mask_taiwan = world_df1['國家'] == "台灣"
world_df_taiwan = world_df1[mask_taiwan]
world_df_taiwan
```

```python
#使用DataFrame的query()方法篩選出
world_df_taiwan = world_df1.query('國家 == "台灣"')
world_df_taiwan
```

```python
#篩選出「國家」欄位為「台灣」且「新增確診數」大於0的資料
taiwan1 = world_df1.query("國家 == '台灣' and 新增確診數 > 0")
taiwan1
```
